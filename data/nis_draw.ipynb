```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draw_nis_chart(filename, percent=95):
    x = []
    y = []
    with open(filename) as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        for row in reader:
            x.append(float(row[0])/1e6)
            y.append(float(row[1]))
        p95 = np.percentile(y, 95)
        p05 = np.percentile(y, 5)
        print('5% percentile {}; 95% percentile: {}'.format(p95, p05))
        
        fig=plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
        plt.plot(x, y)
        plt.plot(x, [p95]*len(y))
        plt.plot(x, [p05]*len(y))
        plt.show()
```

```python
draw_nis_chart('nis_laser.csv')
```

```python
draw_nis_chart('nis_radar.csv')
```
